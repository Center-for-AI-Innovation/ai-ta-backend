````python
# Create a single column that contains the entire drug description. 
# Create drug_description_str by concatenating non-null columns
columns_to_concat = ['description', 'half_life', 'indication', 'mechanism_of_action', 
                    'protein_binding', 'pharmacodynamics', 'state', 'atc_1', 'atc_2',
                    'atc_3', 'atc_4', 'category', 'group', 'pathway', 'molecular_weight',
                    'tpsa', 'clogp']

def create_description(row):
    description = []
    for col in columns_to_concat:
        if pd.notna(row[col]):
            description.append(f"* {col}: {row[col]}")
    return "\n".join(description)

drug_features['drug_description_str'] = drug_features.apply(create_description, axis=1)
drug_features.head()
````

````python
# Create mapping dictionary for drug descriptions
drug_desc_dict = dict(zip(drug_features['node_index'], drug_features['drug_description_str']))

# Add x_drug_description column
kg['x_drug_description'] = None
drug_mask = kg['x_type'] == 'drug'
kg.loc[drug_mask, 'x_drug_description'] = kg.loc[drug_mask, 'x_index'].map(drug_desc_dict)

# Add y_drug_description column
kg['y_drug_description'] = None 
drug_mask = kg['y_type'] == 'drug'
kg.loc[drug_mask, 'y_drug_description'] = kg.loc[drug_mask, 'y_index'].map(drug_desc_dict)

# Add mondo definitions as before
mondo_def_dict = dict(zip(disease_features['mondo_id'], disease_features['mondo_definition']))

kg['y_mondo_definition'] = None
mondo_mask = kg['y_source'] == 'MONDO'
kg.loc[mondo_mask, 'y_mondo_definition'] = kg.loc[mondo_mask, 'y_id'].map(mondo_def_dict)

kg['x_mondo_definition'] = None
mondo_mask = kg['x_source'] == 'MONDO'
kg.loc[mondo_mask, 'x_mondo_definition'] = kg.loc[mondo_mask, 'x_id'].map(mondo_def_dict)

# Preview results
print("Sample rows from enriched knowledge graph:")
print("\nShape of enriched knowledge graph:", kg.shape)
kg.head()
````

# Create final string representation of the triplets

````python
# # Create triplet_string column by concatenating x_name, y_name, relation with definitions
# kg['final_triplet_string'] = kg.apply(lambda row: f"{row['x_name']} -- {row['relation']} -- {row['y_name']}" + 
#                                (f"\n* {row['x_name']}: {row['x_mondo_definition']}" if pd.notna(row['x_mondo_definition']) else "") +
#                                (f"\n* {row['y_name']}: {row['y_mondo_definition']}" if pd.notna(row['y_mondo_definition']) else "") + 
#                                (f"\n* {row['x_name']}: {row['x_drug_description']}" if pd.notna(row['x_drug_description']) else "") + 
#                                (f"\n* {row['y_name']}: {row['y_drug_description']}" if pd.notna(row['y_drug_description']) else "")
#                                , axis=1)

# kg.head()
````

````python
# Create base triplet string
kg['base_string'] = kg.apply(lambda row: f"{row['x_name']} -- {row['display_relation']} (i.e. {row['relation']}) -- {row['y_name']}", axis=1)

# Helper function to create description string only if definition exists
def create_desc_string(name, definition):
    if pd.notna(definition) and definition is not None and definition != '':
        return f"\n**{name}**: {definition}"
    return ''

# Create description strings only when definitions exist
kg['x_mondo_desc'] = kg.apply(lambda row: create_desc_string(row['x_name'], row['x_mondo_definition']), axis=1)
kg['y_mondo_desc'] = kg.apply(lambda row: create_desc_string(row['y_name'], row['y_mondo_definition']), axis=1)
kg['x_drug_desc'] = kg.apply(lambda row: create_desc_string(row['x_name'], row['x_drug_description']), axis=1)
kg['y_drug_desc'] = kg.apply(lambda row: create_desc_string(row['y_name'], row['y_drug_description']), axis=1)

# Concatenate all strings
kg['final_triplet_string'] = (kg['base_string'] + 
                             kg['x_mondo_desc'] + 
                             kg['y_mondo_desc'] + 
                             kg['x_drug_desc'] + 
                             kg['y_drug_desc'])

# Clean up temporary columns
kg.drop(columns=['base_string', 'x_mondo_desc', 'y_mondo_desc', 
                'x_drug_desc', 'y_drug_desc', 'x_mondo_definition', 
                'y_mondo_definition', 'x_drug_description', 'y_drug_description', 
                'x_index', 'x_id', 'x_type', 'x_source', 'y_index', 'y_id', 'y_type', 'y_source'], inplace=True)

kg.head()
````
